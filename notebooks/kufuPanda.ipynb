# Create Index

```python
import numpy as np
import random
```

```python
#Creating non repetitive values for the index

new_index = np.arange(0, 1000)
# new_index

#Shuffling the index 
random.shuffle(new_index)

#Create a new column for the new_index
df_exams['new_index'] =  new_index
df_exams


#Set new_index column as index 
df_exams.set_index('new_index')

#Sort dataframe for new index 
df_exams.sort_index()
```
