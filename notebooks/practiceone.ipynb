```python
# List of countries
countries = ["india", "United States", "britian", "russia"]
print(countries)

# Slicing (this line was fine, just not printed)
print(countries[0:3])

# List of capitals
capitals = ["ND", "Ws", "LN", "MC"]
print(capitals)

# Single example
print(capitals[0] + " is the capital of " + countries[0])

# Correct looping using zip (BEST way)
for cap, country in zip(capitals, countries):
    print(f"{cap} is the capital of {country}")  

# Using extend (flat list)
countries_extended = countries.copy()
countries_extended.extend(capitals)
print(countries_extended)

# Using append (nested list)
countries_appended = countries.copy()
countries_appended.append(capitals)
print(countries_appended)
```
